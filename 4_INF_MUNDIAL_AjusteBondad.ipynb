# BONDAD EDAD PLAYER

## I. CHI SQUARE

```python
bondad.chi_square_test( 'edad_Player', expected_freq=None)
```

## II. KOLMOGOROV

```python
bondad.kolmogorov_smirnov_test('edad_Player', distribution='norm')
```

## III. ANDERSON DARLING

```python
bondad.anderson_darling_test('edad_Player', distribution='norm')
```

## IV. SHAPIRO

```python
bondad.shapiro_wilk_test('edad_Player')
```

## V. JARQUE BERA

```python
bondad.jarque_bera_test('edad_Player')
```

## VI. COMPARACION DISTRIBUCIONES

```python
bondad.plot_distribution_comparison('edad_Player')
```
